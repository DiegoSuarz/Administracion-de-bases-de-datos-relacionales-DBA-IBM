# Laboratorio Práctico: Monitoreo y Optimización de Sus Bases de Datos en MySQL

En este laboratorio, aprenderás a monitorear y optimizar tu base de datos en MySQL con la ayuda de phpMyAdmin.

# Objetivos

Después de completar este laboratorio, podrás:

1. Mantener la salud y el rendimiento de tu base de datos con phpMyAdmin
2. Identificar la diferencia entre una base de datos no optimizada y una optimizada
3. Optimizar tu base de datos con phpMyAdmin aplicando las mejores prácticas

# Software Utilizado en este Laboratorio

En este laboratorio, utilizarás [MySQL](https://www.mysql.com/). MySQL es un Sistema de Gestión de Bases de Datos Relacional (RDBMS) diseñado para almacenar, manipular y recuperar datos de manera eficiente.

![Captura de pantalla de MySQL](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0231EN-SkillsNetwork/labs/MySQL/Lab%20-%20Monitoring%20and%20Optimizing%20your%20Databases%20in%20MySQL/images/mysql.png)

Para completar este laboratorio, utilizarás el servicio de base de datos relacional MySQL disponible como parte de los Laboratorios de la Red de Habilidades de IBM (SN Labs) en el IDE en la nube. SN Labs es un entorno de laboratorio virtual utilizado en este curso.

# Ejercicio 3: Optimiza Tu Base de Datos

La optimización de bases de datos se refiere a maximizar la velocidad y eficiencia de la recuperación de datos de tu base de datos. Al optimizar tu base de datos, mejoras su rendimiento, lo que también puede mejorar el rendimiento de consultas lentas. Al optimizar tu base de datos, mejorarás la experiencia tanto para ti como para tus usuarios.

Podemos ver esto en práctica con el siguiente escenario:

Imagina que eres el organizador de un torneo deportivo con un total de 120 equipos participantes. Has recopilado tres piezas de información de cada equipo: su número de equipo (único), su código de equipo de tres letras para el sistema y la fecha de creación del equipo.

¿Cómo podemos optimizar una base de datos que contenga esa información? ¡Veamos!

### Tarea A: Crea Tu Tabla

1. Para comenzar, crearemos una base de datos en phpMyAdmin.
    
    En la barra lateral izquierda, selecciona el botón **Nuevo** para crear una nueva base de datos.
    
    ![Destacando el botón Nueva Base de Datos en phpMyAdmin](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0231EN-SkillsNetwork/labs/MySQL/Lab%20-%20Monitoring%20and%20Optimizing%20your%20Databases%20in%20MySQL/images/c-create_new_database.png)
    
2. En la página de **Bases de Datos**, puedes ingresar un nombre para la base de datos y dejarla con la codificación **UTF-8**. UTF-8 es la codificación de caracteres más comúnmente utilizada para contenido o datos.
    
    Llamemos a nuestra base de datos **tournament\_teams**.
    
    Presiona **Crear**.
    
    ![Crear una nueva base de datos llamada "tournament_teams" en phpMyAdmin](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0231EN-SkillsNetwork/labs/MySQL/Lab%20-%20Monitoring%20and%20Optimizing%20your%20Databases%20in%20MySQL/images/c-create_tournament_teams.png)
    
3. Ahora, podemos crear una tabla. Llamémosla **teams**.
    
    Recuerda la información que queremos almacenar en esta base de datos: número de equipo, código de equipo y fecha de creación del equipo. Podemos hacer tres columnas para eso.
    
    ![Crear tabla "teams"](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0231EN-SkillsNetwork/labs/MySQL/Lab%20-%20Monitoring%20and%20Optimizing%20your%20Databases%20in%20MySQL/images/c-create_teams_table.png)
    
    Presiona **Continuar**.
    
4. Ahora podemos agregar los nombres de las columnas y los tipos de datos.
    
    Ingresaremos la siguiente información en la tabla. A continuación, solo se muestran los valores que necesitarás completar. Cualquier otra cosa puede dejarse tal como está:
    
    | Nombre | Tipo | Longitud/Valores |
    | --- | --- | --- |
    | team\_no | INT |  |
    | team\_code | CHAR | 100 |
    | creation\_date | CHAR | 100 |
    
    Dado que el número de equipo es único, con cada uno de los 120 equipos teniendo su propio número que varía de 1 a 120, ese será nuestra clave primaria. Lo configuraremos más adelante. Es un tipo de dato numérico, así que usamos **INT** por ahora.
    
    El código de equipo y la fecha de creación se establecen como tipo de dato **CHAR**. Observa que para el tipo **CHAR**, debemos incluir la longitud del valor. Para estar seguros, lo hemos establecido en 100.
    
    Puede que estés pensando que estos tipos de datos no son las mejores opciones para esta base de datos. ¡Tendrías razón! Dado lo que sabemos, hay mejores opciones para cada una de las entradas, pero primero echemos un vistazo a una base de datos no optimizada y cómo podemos optimizarla.
