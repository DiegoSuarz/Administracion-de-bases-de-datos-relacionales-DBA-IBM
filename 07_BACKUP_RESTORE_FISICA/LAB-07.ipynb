### Comprender la Importancia de las Copias de Seguridad Físicas en la Gestión de Bases de Datos.

Una copia de seguridad física implica duplicar todos los componentes de almacenamiento físico, como archivos y directorios, de una tabla específica, base de datos u otros objetos dentro de un sistema de base de datos. Estos componentes incluyen elementos esenciales como archivos de datos, archivos de configuración y archivos de registro, capturando toda la estructura de almacenamiento tal como existe en el disco.

En comparación con las copias de seguridad lógicas, que implican exportar datos en un formato estructurado, las copias de seguridad físicas son generalmente más compactas y rápidas de ejecutar. Esto las hace particularmente adecuadas para escenarios donde bases de datos grandes o críticas deben ser respaldadas de manera rápida y eficiente.

#### Técnicas Comunes para Copias de Seguridad Físicas:

Esta sección explora varios métodos y técnicas utilizados para realizar copias de seguridad físicas en sistemas de bases de datos.

**Copias de Seguridad Basadas en Archivos**: Esta técnica implica copiar archivos de base de datos directamente desde el disco al almacenamiento de respaldo. Es un enfoque sencillo, pero puede requerir tiempo de inactividad o bloqueo de la base de datos durante las operaciones de respaldo. Esto típicamente incluye archivos de datos, archivos de control y registros de redo archivados.

Ejemplo: En un entorno de base de datos Oracle, una copia de seguridad basada en archivos podría implicar copiar los archivos de datos **(.dbf), archivos de control (controlfile.ctl) y registros de redo archivados (\*.arc)** a un destino de respaldo utilizando utilidades del sistema operativo como **cp (copy) o rsync**.

**Copias de Seguridad a Nivel de Bloque:** Las copias de seguridad a nivel de bloque operan a un nivel más bajo, capturando datos a nivel de bloque de disco en lugar de a nivel de archivo. Esta técnica permite copias de seguridad incrementales, donde solo se respaldan los bloques cambiados, reduciendo el tiempo de respaldo y los requisitos de almacenamiento.

Ejemplo: En una base de datos PostgreSQL, las copias de seguridad a nivel de bloque se pueden lograr utilizando herramientas como pg\_basebackup o soluciones de terceros que utilizan el mecanismo de registro anticipado de escritura (WAL) de PostgreSQL para capturar bloques cambiados desde la última copia de seguridad.

**Copias de Seguridad en Línea:** Las copias de seguridad en línea permiten realizar copias de seguridad de la base de datos mientras esta permanece operativa, minimizando el tiempo de inactividad y las interrupciones en las operaciones comerciales.

Ejemplo: En una base de datos MySQL, las copias de seguridad en línea se pueden lograr utilizando herramientas como **mysqldump** para copias de seguridad lógicas o **Percona XtraBackup** para copias de seguridad físicas. Estas herramientas permiten realizar copias de seguridad sin bloquear tablas, asegurando que la base de datos permanezca accesible para los usuarios durante el proceso de respaldo.

**Copias de Seguridad por Instantánea:** Las copias de seguridad por instantánea utilizan instantáneas de almacenamiento para crear copias de la base de datos en un momento dado. Esta técnica proporciona una vista consistente de la base de datos sin interrumpir las transacciones en curso.

Este enfoque a veces se utiliza en la nube tomando instantáneas de volúmenes de almacenamiento basados en la nube y reiniciando la base de datos en una VM diferente.

Ejemplo: En un entorno de Microsoft SQL Server, las copias de seguridad por instantánea se pueden crear utilizando tecnologías de almacenamiento como Storage Spaces Direct (S2D) o soluciones de instantáneas basadas en SAN/NAS. Estas tecnologías aprovechan la infraestructura de almacenamiento para crear instantáneas instantáneas y eficientes en espacio de los volúmenes de la base de datos.

**Compresión de Copias de Seguridad:** Comprimir archivos de respaldo reduce los requisitos de espacio de almacenamiento y mejora la eficiencia de las copias de seguridad sin sacrificar la integridad de los datos.

Ejemplo: En una base de datos MongoDB, se puede habilitar la compresión de copias de seguridad utilizando opciones de compresión integradas en utilidades de respaldo como mongodump o utilizando soluciones de respaldo de terceros que ofrecen características de compresión.

#### Mejores Prácticas para Copias de Seguridad Físicas:

Esta sección ofrece recomendaciones y pautas para implementar y gestionar estrategias de copias de seguridad físicas de manera efectiva.

**Programa de Copias de Seguridad Regular:** Aquí se establece un programa de copias de seguridad regular basado en los requisitos del negocio y la sensibilidad de los datos. Considera factores como la volatilidad de los datos, los objetivos de tiempo de recuperación (RTO) y los objetivos de punto de recuperación (RPO).

**Almacenamiento Externo:** Almacena copias de seguridad en ubicaciones seguras y externas para protegerse contra desastres en el sitio, como incendios, inundaciones o robos.

**Verificación de Copias de Seguridad:** Prueba regularmente los archivos de copia de seguridad para asegurarte de que son archivos válidos, lo que permite una restauración exitosa. La verificación ayuda a identificar cualquier problema con los procesos de copia de seguridad o los medios de almacenamiento antes de que se conviertan en críticos.

**Políticas de Retención:** Define políticas de retención para los archivos de copia de seguridad basadas en requisitos de cumplimiento, estándares regulatorios y necesidades del negocio. Mantén las copias de seguridad durante un período apropiado para apoyar la recuperación de datos y las obligaciones legales.

**Cifrado:** Implementa cifrado para los datos de copia de seguridad en tránsito así como en reposo para proteger la información sensible de accesos no autorizados o violaciones de datos.

**Monitoreo y Alertas:** Aquí se establecen mecanismos de monitoreo y alertas para rastrear el rendimiento de las copias de seguridad, el uso de almacenamiento y posibles problemas. El monitoreo proactivo ayuda a identificar y abordar fallos o anomalías en las copias de seguridad de manera oportuna.

#### Una visión general de los comandos clave de Linux utilizados para realizar copias de seguridad físicas y restauración de la base de datos MySQL

Exploraremos un escenario en el que se te encarga realizar una copia de seguridad física completa del servidor MySQL que alberga una base de datos llamada **world**. Dentro de esta base de datos, encontrarás tablas como **cities**, **countries** y **languages**, cada una conteniendo datos relacionados con **cities**, **countries** y **languages** hablados en todo el mundo.  
Las declaraciones SQL necesarias relacionadas con estos datos se encuentran en un archivo de script sql **world\_mysql\_script.sql**.

Primero puedes descargar este script en la terminal de linux ejecutando el siguiente comando:

```bash
wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0231EN-SkillsNetwork/datasets/World/world_mysql_script.sql
```

Una vez ejecutado, este comando descargará el archivo de script SQL de Internet y lo guardará en el directorio actual de su máquina local.

**Requisito previo:** Para interactuar con bases de datos MySQL en su máquina local, necesita instalar y iniciar el servidor MySQL. Para instalar el servidor MySQL, ejecute el siguiente comando:

```bash
sudo apt-get install mysql-server
```

Este comando te pedirá que ingreses tu contraseña, y una vez que confirmes, descargará e instalará el paquete del servidor MySQL junto con cualquier dependencia necesaria.

> Nota: En caso de que encuentres errores relacionados con paquetes inalcanzables, ejecuta el comando `sudo apt update` para obtener la información más reciente sobre los paquetes disponibles y sus versiones desde los repositorios configurados en tu sistema. Una vez hecho esto, puedes ejecutar el comando para instalar el mysql-server.

Después de la instalación, inicia el servicio de MySQL usando el comando:

```bash
sudo systemctl start mysql
```

Este comando se utiliza para iniciar el servicio de MySQL en un sistema que ejecuta systemd, que es un gestor de sistemas y servicios para Linux.

Aquí está lo que hace cada parte del comando:

- **sudo**: Este prefijo otorga privilegios administrativos temporales al comando que lo sigue. Permite al usuario ejecutar el comando subsiguiente con privilegios de superusuario (root).
    
- **systemctl**: Esta es una utilidad de línea de comandos utilizada para gestionar **systemd**, que es un gestor de sistemas y servicios para sistemas operativos Linux. Permite a los usuarios iniciar, detener, habilitar, deshabilitar y gestionar servicios y otras unidades.
    
- **start**: Este subcomando de **systemctl** instruye a systemd para iniciar el servicio especificado. En este caso, **mysql** se refiere al servicio de MySQL.
    

A continuación, para acceder a la CLI de MySQL (Interfaz de Línea de Comandos), puedes usar el siguiente comando en la terminal:

```bash
sudo mysql -u root -p
```

Aquí está lo que hace cada parte del comando:

- **mysql**: Este es el comando utilizado para lanzar la interfaz de línea de comandos de MySQL.
    
- **\-u root**: Esta opción especifica el nombre de usuario que se utilizará al conectarse al servidor MySQL. En este caso, especifica que se debe usar el usuario root.
    
- **\-p**: Esta opción solicita al usuario que ingrese la contraseña para el usuario de MySQL especificado (en este caso, el usuario root). Después de ingresar la contraseña, si es correcta, el usuario obtiene acceso a la interfaz de línea de comandos de MySQL con privilegios administrativos.
    

Crea la base de datos y navega a la base de datos utilizando los comandos a continuación en la terminal:

```bash
create database world;use world;
```

Utiliza el script **world\_mysql\_script.sql** para completar el proceso de creación de la base de datos world. Este script contiene los datos de las tablas **cities**, **countries** y **languages**.

```bash
source world_mysql_script.sql;
```

Este comando ejecutará todos los comandos SQL contenidos en el archivo de script dentro de la interfaz de línea de comandos de MySQL.

Una vez que la operación esté completa, podemos salir de la interfaz de línea de comandos de MySQL utilizando el comando `\q` en la terminal.

Antes de hacer la copia de seguridad, necesitamos detener el servicio de MySQL usando el comando a continuación:

```bash
sudo systemctl stop mysql
```

Aquí está lo que hace cada parte del comando:

- **systemctl**: Esta es una utilidad de línea de comandos utilizada para gestionar **systemd**, que es un gestor de sistema y servicios para sistemas operativos Linux. Permite a los usuarios iniciar, detener, habilitar, deshabilitar y gestionar servicios y otras unidades.
    
- **stop**: Este subcomando de **systemctl** instruye a systemd para detener el servicio especificado. En este caso, **mysql** se refiere al servicio de MySQL.
    

Luego puedes realizar una copia de seguridad de la base de datos completa utilizando el comando a continuación para copiar recursivamente todos los directorios y subdirectorios del directorio de datos de MySQL al directorio de respaldo especificado:

```bash
sudo cp -R /var/lib/mysql /path/to/backup_directory
```

Aquí está lo que hace cada parte del comando:

- **cp**: Este es el comando utilizado para copiar archivos y directorios.
    
- **\-R**: Esta opción se utiliza para especificar una copia recursiva, lo que significa que copiará el directorio especificado y todo su contenido, incluidos subdirectorios y archivos.
    
- **/var/lib/mysql**: Los archivos de datos de MySQL, incluidas bases de datos, tablas y otros archivos relacionados, se encuentran en este directorio.
    
- **/path/to/backup\_directory**: Este es el directorio de destino donde se copiará el directorio de datos de MySQL. Debes reemplazar “/path/to/backup\_directory” con el directorio donde deseas que se almacene la copia de seguridad.
    

Para verificar la creación exitosa de la copia de seguridad en el directorio de copia de seguridad designado, ejecuta el siguiente comando en la terminal:

```bash
sudo ls -l /path/to/backup_directory
```

Esto enumerará información detallada sobre el contenido del directorio de respaldo especificado y puedes ver los archivos relacionados con la base de datos MySQL aquí.

Puedes restaurar la base de datos MySQL desde el respaldo siguiendo los comandos a continuación:

```bash
sudo mv /var/lib/mysql /var/lib/mysql_old
```

Este comando renombrará el directorio de datos de MySQL actual `/var/lib/mysql` a `/var/lib/mysql_old`

A continuación, podemos copiar recursivamente el contenido del directorio de respaldo que creamos anteriormente en el directorio de datos de MySQL utilizando el siguiente comando:

```bash
sudo cp -R /path/to/backup_directory /var/lib/mysql
```

> Nota: Tenga en cuenta que durante la restauración de la copia de seguridad, puede encontrar espacio insuficiente en el dispositivo. Esto puede ocurrir si el archivo de copia de seguridad o los datos que se están restaurando superan el espacio en disco disponible en el dispositivo donde está instalado MySQL o donde se está restaurando la copia de seguridad. Es esencial asegurarse de que haya suficiente espacio en disco disponible para acomodar el proceso de restauración y evitar encontrar este error.

Podemos cambiar la propiedad del directorio de datos de MySQL y de todo su contenido de manera recursiva al usuario **mysql** y al grupo **mysql** como se indica a continuación:

```bash
sudo chown -R mysql:mysql /var/lib/mysql
```

Aquí está lo que hace cada parte del comando:

- **chown**: Significa **cambiar propiedad**, un comando utilizado para cambiar la propiedad de archivos o directorios en Linux.
    
- **\-R**: Especifica una operación recursiva, incluyendo subdirectorios y archivos.
    
- **mysql:mysql**: Especifica el nuevo propietario y grupo para el directorio y su contenido. En este caso, establece el usuario como “mysql” y el grupo como “mysql”.
    
- **/var/lib/mysql**: Especifica el directorio cuya propiedad será cambiada. Este es típicamente el directorio de datos de MySQL.
    

Una vez completado, puedes reiniciar el servicio de MySQL usando el comando: `sudo systemctl start mysql`
