### Tarea Extra 2.5.1 - Escriba un script bash que realice una copia de seguridad de todas las bases de datos

`mysqldump` es una herramienta de línea de comandos que realiza copias de seguridad lógicas de una base de datos.

Su sintaxis genérica es `mysqldump db_name > backup-file.sql`

Su sintaxis extendida es `mysqldump --all-databases --user=root --password=NzA4Ny1y > backup-file.sql`

Escriba un script bash llamado `mybackup.sh` que realice las siguientes tareas.

- Realice la copia de seguridad de todas las bases de datos usando mysqldump
- Almacene la salida en el archivo all-databases-backup.sql
- En el directorio /tmp, cree un directorio nombrado con la fecha actual como YYYYMMDD. Por ejemplo, **20210830**
- Mueva el archivo all-databases-backup.sql al directorio `/tmp/mysqldumps/<fecha actual>/`

Tome una captura de pantalla del script bash con todo el código claramente visible.
